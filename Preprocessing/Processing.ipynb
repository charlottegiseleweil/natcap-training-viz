```python
import numpy as np
import pandas as pd
import pycountry
import collections
import json

DATA_DIR = '../Data/'
```

# Shikhar + Varun

```python
processed = pd.read_csv(DATA_DIR + 'processed_Data.csv')
processed.head()
```

```python
# DUPLICATES!
processed[processed['Year'] == 2017][processed[processed['Year'] == 2017]['Country'] == 'USA']
```

```python
processed.dropna(inplace=True)
processed.Country = [pycountry.countries.get(alpha_3=iso3).alpha_2 for iso3 in processed.Country]
processed.head()
```

```python
total = processed[['Country', 'Number_of_Participants']].groupby('Country').sum().reset_index()
total.loc[:, 'Year'] = 'Total'
total.head()
```

```python
years = processed[processed['Number_of_Participants'] > 0]
years = years[['Year', 'Country', 'Number_of_Participants']]
years = years.sort_values('Year')
years.head()
```

```python
data = pd.concat([total, years], sort=True)
data = data.set_index(['Year', 'Country'])
data
```

```python
# DUPLICATES!
data.loc[(2018, 'US'), :]
```

```python
nested_dict = collections.defaultdict(dict)

for keys, value in data.Number_of_Participants.iteritems():
    nested_dict[keys[0]][keys[1]] = value
```

```python
with open('data.json', 'w') as outfile:
    json.dump(nested_dict, outfile)
```

```python
### People trained

attendance = pd.DataFrame()

# years 
for yr in data.keys():

    attendance = attendance.append({'Year': yr,

                                    # number of participants (P) for year yr
                                    'Total' : len(data[yr][data[yr]['P/I']=='P']), 

                                    # number of participants for year yr per type
                                    'Intro' : len(data[yr][(data[yr]['P/I']=='P') & (data[yr]['TYPE']=='Intro')]),
                                    'Nodal' : len(data[yr][(data[yr]['P/I']=='P') & (data[yr]['TYPE']=='Nodal')]),
                                    'Partner':len(data[yr][(data[yr]['P/I']=='P') & (data[yr]['TYPE']=='Partner')]),
                                    'Public': len(data[yr][(data[yr]['P/I']=='P') & (data[yr]['TYPE']=='Public')]),
                                    'Regional': len(data[yr][(data[yr]['P/I']=='P') & (data[yr]['TYPE']=='Regional')]),
                                    'Seminar': len(data[yr][(data[yr]['P/I']=='P') & (data[yr]['TYPE']=='Seminar')]),
                                    'Symposium': len(data[yr][(data[yr]['P/I']=='P') & (data[yr]['TYPE']=='Symposium')]),
                                    'Webinar': len(data[yr][(data[yr]['P/I']=='P') & (data[yr]['TYPE']=='Webinar')]),
                                    #'Other training': len(data[yr][(data[yr]['P/I']=='P') & (data[yr]['TYPE']=='nan')])
                                   },
                                   ignore_index=True)

attendance['Other trainings'] = attendance['Total'] - attendance['Symposium']   
attendance = attendance.sort_values('Year')
attendance = attendance.set_index('Year')

# Get totals
attendance = attendance.append(pd.Series(attendance.sum(numeric_only=True), name='Total'), ignore_index=False)

attendance
```
